```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image1 = cv2.imread("q1.png")
image2 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)

kernel = np.ones((11,11), np.float32) / 121
smooth_img = cv2.filter2D(image2, -1, kernel)

kernel2 = np.array([[0,-1,0],
                    [-1,5,-1],
                    [0,-1,0]])
sharpen_img = cv2.filter2D(image2, -1, kernel2)

plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
plt.imshow(image2)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(smooth_img)
plt.title("Smoothed Image")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(sharpen_img)
plt.title("Sharpened Image")
plt.axis("off")
plt.show()

laplacian = cv2.Laplacian(image2, cv2.CV_64F)

plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.imshow(image2)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(np.uint8(np.absolute(laplacian)), cmap='gray')
plt.title("Laplacian Edge Detection")
plt.axis("off")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
image1 = cv2.imread("q1.png")
image2 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
kernel = np.ones((5,5), np.float32) / 25
image3 = cv2.filter2D(image2, -1, kernel)
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(image2)
plt.title("Original Image")
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(image3)
plt.title("Averaging Filter")
plt.axis("off")
plt.show()
kernel1 = np.array([[1,2,1],
                    [2,4,2],
                    [1,2,1]]) / 16
weighted_avg = cv2.filter2D(image2, -1, kernel1)
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(image2)
plt.title("Original Image")
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(weighted_avg)
plt.title("Weighted Average Filter")
plt.axis("off")
plt.show()
gaussian_blur = cv2.GaussianBlur(image2, (5,5), 0)
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(image2)
plt.title("Original Image")
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(gaussian_blur)
plt.title("Gaussian Blur")
plt.axis("off")
plt.show()
median = cv2.medianBlur(image2, 5)
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(image2)
plt.title("Original Image")
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(median)
plt.title("Median Filter")
plt.axis("off")
plt.show()
```
